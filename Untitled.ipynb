# Plotting Script to over plot param recovery from cadence tests

```python
def openPickle(dir):
	file = dir
	t_pkl = open(file, 'rb')
	params = pickle.load(t_pkl )
	t_pkl.close()
	#print timescale[:, 10,  100]
	return params
```

```python
import matplotlib.pyplot as plt
from matplotlib import*
import pandas
%matplotlib inline

zwick0 = np.loadtxt('kplr006932990Carini-calibrated.dat',delimiter=' ')
newcols = [ 'cadence','mask' ,'cal_time' ,'y' ,'y_err','t',' time_err']	
zwickdata = pandas.DataFrame(data=zwick0,columns=newcols, dtype=float)
t, y, mask = zwickdata['t'],zwickdata['y'],zwickdata['mask']
w = np.where(mask == 1)[0]


import seaborn as sns
sns.set_palette("colorblind")
sns.set_style("white")
#set width and height in inches you want 
plt.figure(figsize=(10,7))
plt.scatter(t[w],y[w],color ='k',marker = "+",s =1, label = 'Kepler AGN Zwicky 229.15' )
plt.ylabel('Optical Flux [e/s]',fontsize=15)
plt.xlabel('time [days]',fontsize=15)
plt.ylim([5.3e-7,8.2e-07])
plt.legend(loc="upper right", fontsize=18,markerscale=0) 
#dpi is resolution set high for posters
plt.savefig('exampleKeplerAGN.png', dpi = 100)
```

```python
import copy

def samplesPlot(dir):
    coeff = openPickle(dir)
    ndims = coeff.shape[0]
    nwalkers = coeff.shape[1]
    nsteps = coeff.shape[2]
    samples = np.swapaxes(copy.copy( coeff[:, :, nsteps/2:]).reshape((ndims, -1), order='F'), axis1=0, axis2=1)
    return samples
samples = samplesPlot(dir)
```

```python
fig = corner.corner(samples, labels=[r"$\alpha_1$", r"$a_2$", "$b_0$","$b_1$"],
                      truths=[0.75, 0.01,7.0e-9,1.2e-9])
fig.savefig("TESTtriangle.png")
```
